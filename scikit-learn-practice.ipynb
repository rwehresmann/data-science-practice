```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
heart_disease = pd.read_csv('data/heart-disease.csv')
```

```python
x = heart_disease.drop('target', axis=1)
y = heart_disease['target']
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100)

clf.get_params()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
clf.fit(x_train, y_train)
```

```python
x_train
```

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
clf.score(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f"Model accuracy on test set: {clf.score(x_test, y_test)}")
```

```python
import pickle

pickle.dump(clf, open('models/random_forest_model1_1.pkl', 'wb'))
```

```python
loaded_model = pickle.load(open('models/random_forest_model1_1.pkl', 'rb'))
loaded_model.score(x_test, y_test)
```

## Dealing with missing data

```python
car_sales = pd.read_csv('data/car-sales-extended-missing-data.csv')
```

```python
car_sales.Make.fillna('missing', inplace=True)

car_sales.Colour.fillna('missing', inplace=True)

car_sales['Odometer (KM)'].fillna(car_sales['Odometer (KM)'].mean(), inplace=True)

car_sales.Doors.fillna(4, inplace=True)
```

```python
car_sales.isna().sum()
```

```python
car_sales.dropna(inplace=True)
```

```python
x = car_sales.drop('Price', axis=1)
y = car_sales.Price
```
